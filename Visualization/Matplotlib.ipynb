```python
x=np.random.rand(50)
y=np.random.rand(50)
x,y
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,c='r', alpha = 0.4,marker='x')
#c is for color and search html color code for different code
#alpha == opacity
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title('title')
plt.grid()
```

```python
x=np.linspace(1,10,100)
y=np.sin(x)
```

```python
#plt.scatter(x,y)
plt.figure(figsize=(8,2))
plt.plot(x,y,'--r')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title('sin')
plt.grid()
```
