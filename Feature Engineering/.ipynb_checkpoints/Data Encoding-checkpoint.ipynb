```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({
    'color': ['red','blue','green','green','red','blue']
})
```

```python
df.head()
```

```python
##create instance of OneHotEncoder
encoder = OneHotEncoder()
```

```python
#encoder.fit(df[['color']])
#encoder.transform(df[['color']])
encoder.fit_transform(df[['color']])
```

```python
encoded = encoder.fit_transform(df[['color']]).toarray()
```

```python
encoder_df = pd.DataFrame(encoded,columns=encoder.get_feature_names_out())
```

```python
encoder_df
```

```python
#for any new value
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df,encoder_df],axis=1)
```

```python
#on tips data convert all categorical features to nominal vector using OHE
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder_tips = OneHotEncoder()
```

```python
tips_encoded = pd.DataFrame(encoder_tips.fit_transform(tips[['sex','smoker','day','time']]).toarray(),columns=encoder_tips.get_feature_names_out())
```

```python
pd.concat([tips,tips_encoded],axis=1)
```
