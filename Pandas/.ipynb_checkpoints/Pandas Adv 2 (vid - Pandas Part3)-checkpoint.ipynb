```python
df =  pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df.sample()
```

```python
df.groupby('Survived').mean()
df.groupby('Survived').sum()
```

```python
df.groupby('Sex').sum()
df.groupby('Sex').sum()['Fare']
```

```python
df.groupby('Pclass').mean()
df.groupby('Pclass').sum()
df.groupby('Pclass').sum()['Fare']
```

```python
Pclass = df.groupby('Pclass').sum()
Pclass.T, Pclass.transpose()
```

```python
df21 = pd.DataFrame({'key1':[1,2,4,5,6],
                   'key2':[4,5,6,7,8],
                   'key3':[3,4,5,6,6]})
df22 = pd.DataFrame({'key1':[1,2,45,6,67],
                   'key4':[56,5,6,7,8],
                   'key5':[3,56,5,6,6]})
df23 = pd.DataFrame({'key1':[1,2,45,6,67],
                   'key2':[56,5,6,7,8],
                   'key5':[3,56,5,6,6]})
df21,df22
```

```python
#default inner join based on common column
pd.merge(df21,df22)
pd.merge(df21,df22,how='left')
pd.merge(df21,df22,how='right')
pd.merge(df21,df22,how='outer')
```
