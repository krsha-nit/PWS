```python
df21 = pd.DataFrame({'key1':[1,2,4,5,6],
                   'key2':[4,5,6,7,8],
                   'key3':[3,4,5,6,6]})
df22 = pd.DataFrame({'key1':[1,2,45,6,67],
                   'key4':[56,5,6,7,8],
                   'key5':[3,56,5,6,6]})
df23 = pd.DataFrame({'key1':[1,2,45,6,67],
                   'key2':[56,5,6,7,8],
                   'key5':[3,56,5,6,6]})
df21,df22
```

```python
#default inner join based on common column
pd.merge(df21,df22)
pd.merge(df21,df22,how='left')
pd.merge(df21,df22,how='right')
pd.merge(df21,df22,how='outer')
pd.merge(df21,df22,how='cross') #cross join -- table1*table2
```

```python
#takes all the common columns as key for joining by default
pd.merge(df21,df23)
pd.merge(df21,df23,on='key1')
```

```python
#if different keys in 2 tables
pd.merge(df21,df23,how = 'left',left_on='key2',right_on='key5')
```

```python
df31 = pd.DataFrame({'key1':[1,2,4,5,6],
                   'key2':[4,5,6,7,8],
                   'key3':[3,4,5,6,6]},
                    index = ['a','b','c','d','e']
                   )
df32 = pd.DataFrame({'key11':[1,2,45,6,67],
                   'key22':[56,5,6,7,8],
                   'key33':[3,56,5,6,6]},
                   index = ['a','b','h','i','j'])
df31,df32
```

```python
#join -- default left join based on index
df31.join(df32)
df31.join(df32,how='inner')
df31.join(df32,how='right')
df31.join(df32,how='outer')
df31.join(df32,how='cross') #table1*table2
```

```python
#external function
def fare_flag(x):
    if x<10:
        return 'low'
    elif x>=10 and x<20:
        return 'mid'
    else:
        return 'high'
    
df['fare_flag']=df['Fare'].apply(fare_flag)
df.head()
```
