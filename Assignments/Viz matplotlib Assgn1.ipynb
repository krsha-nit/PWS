#### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.
    Ans. Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations in Python. It provides a wide range of functions and methods for creating various types of plots and charts. Matplotlib is widely used in the scientific community, data analysis, and visualization tasks due to its flexibility, ease of use, and extensive customization options.

    1. Line Plot: A line plot is used to visualize the relationship between two variables. It connects individual data points with straight lines, making it useful for displaying trends over time or continuous data.

    2. Scatter Plot: A scatter plot displays individual data points as dots on a two-dimensional plane. It is useful for visualizing the relationship between two variables and identifying patterns or correlations.

    3. Bar Plot: A bar plot (or bar chart) represents categorical data using rectangular bars of varying heights or lengths. It is commonly used to compare and display the distribution of data across different categories.

    4. Histogram: A histogram is used to visualize the distribution of a single variable by dividing the data into bins and displaying the frequency or count of values within each bin. It provides insights into the underlying data distribution and helps identify patterns or outliers.

    5. Pie Chart: A pie chart represents data as a circular plot divided into sectors, with each sector representing a specific category or value proportion. It is effective for displaying the composition or relative proportions of different parts of a whole.

#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated dataplot a scatter plot.
    import numpy as np
    np.random.seed(3)
    x = 3 + np.random.normal(0, 2, 50)
    y = 3 + np.random.normal(0, 2, len(x))
    Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.figure(figsize=(10,6))
#plt.plot(x,y)
plt.scatter(x,y)
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.title('title')
```
