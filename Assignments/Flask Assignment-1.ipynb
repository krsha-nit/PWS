#### Q1. What is Flask Framework? What are the advantages of Flask Framework?
    Ans. Flask is a micro web framework for Python. It is designed to be lightweight, simple, and easy to use. Flask allows developers to build web applications quickly and with minimal boilerplate code. It is often referred to as a "micro" framework because it does not impose strict rules or dependencies on how to build applications. Instead, it provides the basic tools needed to build a web application and leaves the rest to the developer's choice.

    Advantages of Flask Framework:

    Lightweight: Flask is a minimalistic framework, making it easy to learn and understand. It allows developers to build applications without unnecessary overhead.
    Flexibility: Flask gives developers the freedom to choose their preferred tools and libraries for tasks like database management, authentication, and more.
    Extensibility: Flask can be easily extended with various Flask extensions to add extra functionality when needed.
    Built-in Development Server: Flask comes with a built-in development server, so developers can test and run their applications without requiring additional setup.
    Easy to Get Started: Flask's simplicity and intuitive design make it an excellent choice for beginners and small projects.
    Large Community and Documentation: Flask has a vibrant community, providing extensive documentation, tutorials, and resources for developers.

#### Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

```python
# Import the Flask module
from flask import Flask

# Create a Flask application
app = Flask(__name__)

# Define a route for the home page
@app.route('/')
def hello_world():
    return 'Hello World!!'

# Run the application
if __name__ == '__main__':
    app.run()
```

#### Q3. What is App routing in Flask? Why do we use app routes?
    Ans. In Flask, app routing refers to the process of mapping URLs (Uniform Resource Locators) to the functions that handle them. It allows developers to define different URL patterns and the corresponding actions that should be taken when a user accesses those URLs. This is achieved using the @app.route() decorator in Flask.
    
    @app.route('/')
    def index():
        return 'This is the home page.'

    @app.route('/about')
    def about():
        return 'This is the about page.'

    In the above code, when a user visits the root URL /, the index() function will be executed and return 'This is the home page.' Similarly, when a user visits the /about URL, the about() function will be executed and return 'This is the about page.'
    
    
    Usage of app routes:
        App routing in Flask allows developers to define the structure of their web application and handle different user requests based on the URLs accessed.
        It provides a clean and organized way to manage the different endpoints of the application and the corresponding actions.
        By using app routes, developers can create RESTful APIs, web pages, or any other web-based application with ease and maintainability.
        App routes enable the application to serve specific content or perform specific tasks based on user interactions, making the web application dynamic and interactive.

#### Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
    Company Name: ABC Corporation
    Location: India
    Contact Detail: 999-999-9999
    Attach the screenshot of the output in Jupyter Notebook.

```python
from flask import Flask

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def company_details():
    return '''Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999'''

if __name__ == '__main__':
    app.run()
```

#### Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.
    Ans. In Flask, the url_for() function is used for URL building. It generates a URL for the specified endpoint (view function) based on its name and any additional arguments provided. This function is helpful because it allows you to change the URL pattern in the future without affecting the code that uses the function to build URLs. Example below:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/welcome')
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def company_details():
    return f'''Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999
<a href="{url_for('welcome')}">Go to Welcome Page</a>'''

if __name__ == '__main__':
    app.run()
```
