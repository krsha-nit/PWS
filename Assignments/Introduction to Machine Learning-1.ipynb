#### Q2. What is supervised learning? List some examples of supervised learning.
    Ans. Supervised learning is a type of machine learning where the algorithm learns from labeled training data. In supervised learning, each example in the training dataset has a corresponding label or target variable that the algorithm tries to predict. The algorithm learns to map input features to the correct output based on the given labels, and its performance is evaluated on its ability to make accurate predictions on new, unseen data.

    Examples of Supervised Learning:

    Image Classification: Given a dataset of images and their corresponding labels (e.g., cat, dog, car), the algorithm learns to classify new images into the correct categories.

    Sentiment Analysis: In this case, the algorithm is trained on a dataset of text samples labeled with sentiment (e.g., positive, negative, neutral), and it can predict the sentiment of new, unseen text.

    Regression: Supervised learning can also be used for regression tasks, such as predicting a numerical value. For instance, given historical data on house prices and relevant features (e.g., area, number of bedrooms), the algorithm can predict the price of a new house.

    Speech Recognition: With a dataset of audio samples and corresponding transcriptions, a supervised learning algorithm can learn to convert spoken words into text.

    Medical Diagnosis: Using historical medical records with diagnosed conditions, a supervised learning model can be built to assist in diagnosing new patients based on their symptoms and test results.

#### Q3. What is unsupervised learning? List some examples of unsupervised learning.
    Ans. Unsupervised learning is a type of machine learning where the algorithm learns from unlabeled data, without specific guidance in the form of target variables. The goal of unsupervised learning is to find patterns, structures, or representations within the data without being explicitly told what to look for.

    Examples of Unsupervised Learning:

    Clustering: Unsupervised learning is commonly used for clustering similar data points together. For example, clustering can be applied to customer segmentation, grouping similar customers based on their behavior or preferences.

    Anomaly Detection: In this case, the algorithm learns to identify unusual or rare data points that differ significantly from the rest of the data. Anomaly detection is used in fraud detection and identifying faulty components in manufacturing.

    Dimensionality Reduction: Unsupervised learning techniques like Principal Component Analysis (PCA) and t-distributed Stochastic Neighbor Embedding (t-SNE) can reduce the number of features in high-dimensional data while preserving important patterns.

    Topic Modeling: Unsupervised learning can be used to extract underlying topics from a collection of text documents without any prior knowledge of what those topics might be.

    Recommendation Systems: Unsupervised learning techniques can be applied to recommend products, movies, or content to users based on their past behaviors or preferences.

#### Q5. What are the main differences between supervised, unsupervised and semi-supervised learning?
    Ans. 
    Supervised Learning:
    In supervised learning, the algorithm is trained on labeled data, where each example in the dataset has a corresponding target or label.
    The goal is to learn a mapping from input features to output labels so that the algorithm can make predictions on new, unseen data.
    Supervised learning is used for tasks like classification and regression, where the model learns to predict discrete classes or continuous values.
    
    Unsupervised Learning:
    In unsupervised learning, the algorithm is trained on unlabeled data, and there are no explicit target variables provided during training.
    The goal is to find patterns, structures, or representations within the data without any specific guidance.
    Unsupervised learning is used for tasks like clustering, where similar data points are grouped together based on their inherent similarities, and dimensionality reduction, which reduces the number of features while preserving important patterns.
    
    Semi-Supervised Learning:
    Semi-supervised learning is a combination of supervised and unsupervised learning approaches.
    In this approach, the algorithm is trained on a small amount of labeled data along with a larger amount of unlabeled data.
    The idea is that the unlabeled data can provide additional information to improve the performance of the model, especially when labeled data is limited or expensive to obtain.
    Semi-supervised learning can be beneficial in cases where obtaining a fully labeled dataset is challenging.

#### Q6. What is train, test and validation split? Explain the importance of each term.
    Ans. Train, test, and validation split is a common practice in machine learning and deep learning workflows to evaluate and validate the performance of a model during training and ensure its generalizability on unseen data. The dataset is divided into three subsets:

    Training Set:
    The training set is the largest portion of the dataset and is used to train the machine learning model.
    The model learns from the patterns and relationships present in the training data.
    A well-trained model can generalize these patterns to make accurate predictions on new, unseen data.
    
    Test Set:
    The test set is a separate portion of the dataset that the model has not seen during training.
    It is used to evaluate the model's performance and assess how well it generalizes to new data.
    The test set provides an unbiased estimate of the model's accuracy and helps identify potential overfitting or underfitting issues.
    
    Validation Set:
    The validation set is an optional subset of the dataset that is used for hyperparameter tuning and model selection.
    During training, the model's performance is monitored on the validation set to fine-tune parameters and avoid overfitting.
    The validation set acts as a proxy for the unseen data, helping to choose the best-performing model before applying it to the test set.
    
    
    The importance of each term:

        Training Set: The training set is essential because it is used to teach the model the underlying patterns and relationships in the data. The model's ability to learn from the training data directly impacts its performance on unseen data.

        Test Set: The test set is critical to assess the model's generalization capabilities. It gives an estimate of how well the model will perform on real-world, unseen data. The test set provides an unbiased evaluation of the model's performance.

        Validation Set: The validation set is crucial for fine-tuning the model and preventing overfitting. It helps to choose the best model configuration and hyperparameters based on performance on the unseen validation data.

    By splitting the dataset into these subsets, we ensure that the model's evaluation is independent of the data it was trained on and that it can make accurate predictions on new, unseen data.

#### Q7. How can unsupervised learning be used in anomaly detection?
    Ans. Unsupervised learning can be used in anomaly detection to identify unusual patterns or outliers in data without the need for labeled examples of anomalies. The key idea is that anomalies are typically rare and distinct from the majority of the data, making them stand out as patterns that do not conform to the normal behavior observed in the majority of the data.

    The process of using unsupervised learning for anomaly detection typically involves the following steps:

        Data Representation: The data is first represented in a suitable format for unsupervised learning. This may involve feature engineering or data preprocessing to extract relevant information and reduce noise.

        Unsupervised Learning Algorithm: An unsupervised learning algorithm, such as clustering or density estimation, is applied to the data to model the normal behavior or patterns in the data.

        Anomaly Scoring: Once the unsupervised model is trained on the majority of the data, it can assign anomaly scores to new data points. These scores indicate the deviation of a data point from the learned normal patterns.

        Thresholding: A threshold is set to distinguish between normal and anomalous data points. Data points with anomaly scores above the threshold are flagged as anomalies.

        Evaluation: The performance of the anomaly detection model is evaluated on labeled anomaly data, if available. Common evaluation metrics include precision, recall, and F1-score.

    Unsupervised learning methods used for anomaly detection include:

        Clustering: Anomalous data points may form separate clusters or be far away from the main clusters in the feature space.

        Density Estimation: Anomalous data points may have low density regions, making them distinguishable from regions with high data density.

        Autoencoders: Autoencoders are neural network architectures used for unsupervised feature learning. They can reconstruct normal data points well, but anomalies may have larger reconstruction errors.

    Unsupervised anomaly detection is particularly useful in scenarios where labeled anomaly data is scarce or difficult to obtain. It allows for the detection of previously unknown anomalies, making it valuable in various applications, including fraud detection, intrusion detection, and fault detection in industrial systems.

#### Q8. List down some commonly used supervised learning algorithms and unsupervised learning algorithms.
    Ans. Commonly Used Supervised Learning Algorithms:

        Linear Regression: Used for regression tasks, it models the relationship between dependent and independent variables as a linear equation.

        Logistic Regression: Used for binary classification tasks, it predicts the probability of an instance belonging to a particular class.

        Decision Trees: Tree-like models used for both classification and regression tasks, making decisions based on the features' values.

        Random Forest: An ensemble method combining multiple decision trees to improve performance and reduce overfitting.

        Support Vector Machines (SVM): Effective for both classification and regression, SVM finds the hyperplane that best separates data points of different classes.

        K-Nearest Neighbors (KNN): A simple algorithm that classifies data points based on the majority class among their k-nearest neighbors.

        Gradient Boosting Machines (GBM): Another ensemble method that combines weak learners (usually decision trees) to create a powerful model.

        Naive Bayes: Based on Bayes' theorem, it is used for classification tasks and is particularly effective with text data and spam filtering.

        Neural Networks: Deep learning models composed of multiple layers of interconnected nodes, used for complex tasks like image recognition and natural language processing.

    Commonly Used Unsupervised Learning Algorithms:

        K-Means Clustering: Divides data into k clusters based on similarity, aiming to minimize the intra-cluster distance and maximize inter-cluster distance.

        Hierarchical Clustering: Builds a tree-like hierarchy of clusters to group similar data points at different levels of granularity.

        DBSCAN (Density-Based Spatial Clustering of Applications with Noise): Identifies clusters of arbitrary shapes based on data density.

        Gaussian Mixture Model (GMM): A probabilistic model that assumes data points are generated from a mixture of several Gaussian distributions.

        Principal Component Analysis (PCA): A dimensionality reduction technique that projects data onto a lower-dimensional space while preserving most of its variance.

        t-distributed Stochastic Neighbor Embedding (t-SNE): Another dimensionality reduction method that emphasizes preserving local structures in high-dimensional data.

        Autoencoders: Neural networks used for unsupervised feature learning and dimensionality reduction.

        Anomaly Detection Algorithms: Various unsupervised methods for identifying anomalies or outliers in data, such as isolation forests and one-class SVM.
