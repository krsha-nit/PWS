### Q1. What is the mathematical formula for a linear SVM?
Ans. In a binary classification setting, a linear Support Vector Machine (SVM) finds the optimal hyperplane that separates two classes. The mathematical formula for a linear SVM is as follows:

Given a training dataset with input features xᵢ (i = 1, 2, ..., n) and corresponding class labels yᵢ (yᵢ ∈ {-1, +1}), the objective of a linear SVM is to find a weight vector w and bias term b that defines the hyperplane equation:

f(x) = w^T * x + b

where:

w is the weight vector (normal to the hyperplane).
x is the input feature vector.
b is the bias term (intercept).
The SVM tries to maximize the margin between the two classes while minimizing the classification error. The margin is the distance between the hyperplane and the closest data points (support vectors) of both classes.


### Q2. What is the objective function of a linear SVM?
Ans. The objective function of a linear SVM aims to find the optimal hyperplane that maximizes the margin while also considering the classification accuracy on the training data. This can be formulated as an optimization problem, where we minimize the squared norm of the weight vector (to maximize the margin) subject to the constraint that all data points are classified correctly:

Minimize: 1/2 * ||w||²

Subject to: yᵢ * (w^T * xᵢ + b) ≥ 1 for all data points (xᵢ, yᵢ)

Here, ||w||² represents the squared L2-norm of the weight vector w.


### Q3. What is the kernel trick in SVM?
Ans. The kernel trick is a technique used in Support Vector Machines to extend the linear SVM to non-linear decision boundaries. It allows SVMs to efficiently handle data that may not be linearly separable in the original feature space by implicitly mapping the data into a higher-dimensional space where it might become linearly separable.

Mathematically, instead of computing the dot product of the input feature vectors directly, the kernel function K(xᵢ, xⱼ) is used, which represents the dot product of their mapped representations in the higher-dimensional space:

K(xᵢ, xⱼ) = φ(xᵢ) ⋅ φ(xⱼ)

Here, φ(x) represents the mapping of the input x into the higher-dimensional space. The kernel function is chosen so that the computations in the higher-dimensional space are avoided explicitly, saving computational resources.

Commonly used kernel functions include the polynomial kernel (K(xᵢ, xⱼ) = (xᵢ ⋅ xⱼ + c)ᵈ) and the radial basis function (RBF) kernel (K(xᵢ, xⱼ) = exp(-γ * ||xᵢ - xⱼ||²)), among others.


### Q4. What is the role of support vectors in SVM Explain with example
Ans. In SVM, support vectors are the data points that lie closest to the decision boundary (hyperplane). These are the critical data points that play a crucial role in defining the decision boundary and, therefore, the classification performance of the SVM. Support vectors are the only data points that influence the parameters of the hyperplane (weights and bias).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Sample data
X = np.array([[1, 2], [2, 3], [2, 5], [3, 2], [3, 4], [4, 3]])
y = np.array([1, 1, 1, -1, -1, -1])

# Fit the SVM classifier
clf = SVC(kernel='linear')
clf.fit(X, y)

# Get support vectors
support_vectors = clf.support_vectors_

# Plot the data points and the decision boundary
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, marker='o', s=100, label='Data Points')
plt.scatter(support_vectors[:, 0], support_vectors[:, 1], facecolors='none', edgecolors='k', s=200, label='Support Vectors')

# Plot the hyperplane (decision boundary)
w = clf.coef_[0]
b = clf.intercept_[0]
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, levels=[-1, 0, 1], linestyles=['--', '-', '--'], colors='k')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.title('Support Vectors in SVM')
plt.show()
```

### Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in SVM?
In SVM, the margin refers to the distance between the hyperplane and the closest data points (support vectors). The objective is to find the hyperplane that maximizes this margin while still correctly classifying the data points.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Sample data
X = np.array([[2, 3], [3, 2], [3, 4], [6, 5], [7, 6], [8, 5]])
y = np.array([1, 1, 1, -1, -1, -1])

# Fit the SVM classifier with a linear kernel (for illustration purposes)
hard_margin_clf = SVC(kernel='linear', C=1000)
hard_margin_clf.fit(X, y)

soft_margin_clf = SVC(kernel='linear', C=0.1)  # Smaller C allows more misclassifications (soft margin)
soft_margin_clf.fit(X, y)

# Get support vectors
hard_margin_support_vectors = hard_margin_clf.support_vectors_
soft_margin_support_vectors = soft_margin_clf.support_vectors_

# Plot the data points and the decision boundaries
plt.figure(figsize=(10, 4))

# Hard Margin SVM
plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, marker='o', s=100, label='Data Points')
plt.scatter(hard_margin_support_vectors[:, 0], hard_margin_support_vectors[:, 1], facecolors='none', edgecolors='k', s=200, label='Support Vectors')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Hard Margin SVM')
w_hard = hard_margin_clf.coef_[0]
b_hard = hard_margin_clf.intercept_[0]
xx, yy = np.meshgrid(np.arange(X[:, 0].min() - 1, X[:, 0].max() + 1, 0.1), np.arange(X[:, 1].min() - 1, X[:, 1].max() + 1, 0.1))
Z = hard_margin_clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, levels=[-1, 0, 1], linestyles=['--', '-', '--'], colors='k')
plt.legend()

# Soft Margin SVM
plt.subplot(122)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, marker='o', s=100, label='Data Points')
plt.scatter(soft_margin_support_vectors[:, 0], soft_margin_support_vectors[:, 1], facecolors='none', edgecolors='k', s=200, label='Support Vectors')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Soft Margin SVM')
w_soft = soft_margin_clf.coef_[0]
b_soft = soft_margin_clf.intercept_[0]
xx, yy = np.meshgrid(np.arange(X[:, 0].min() - 1, X[:, 0].max() + 1, 0.1), np.arange(X[:, 1].min() - 1, X[:, 1].max() + 1, 0.1))
Z = soft_margin_clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
plt.contour(xx, yy, Z, levels=[-1, 0, 1], linestyles=['--', '-', '--'], colors='k')
plt.contour(xx, yy, Z, levels=[-1, 1], colors='gray', alpha=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

### Q6. SVM Implementation through Iris dataset.
~ Load the iris dataset from the scikit-learn library and split it into a training set and a testing setl
~ Train a linear SVM classifier on the training set and predict the labels for the testing setl
~ Compute the accuracy of the model on the testing setl
~ Plot the decision boundaries of the trained model using two of the featuresl
~ Try different values of the regularisation parameter C and see how it affects the performance of
the model.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X, y = iris.data[:, :2], iris.target

# Split the dataset into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a linear SVM classifier on the training set
clf = SVC(kernel='linear')
clf.fit(X_train, y_train)

# Predict the labels for the testing set
y_pred = clf.predict(X_test)

# Compute the accuracy of the model on the testing set
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of the SVM model: {accuracy:.2f}")

# Plot the decision boundaries of the trained model using two features
def plot_decision_boundary(clf, X, y):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.Paired)
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title('Decision Boundaries of the Trained Model')
    plt.show()

# Plot decision boundaries using two features
plot_decision_boundary(clf, X_train, y_train)

# Try different values of the regularization parameter C and observe performance
C_values = [0.01, 0.1, 1, 10, 100]
for C_val in C_values:
    clf = SVC(kernel='linear', C=C_val)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy with C = {C_val}: {accuracy:.2f}")
```

Bonus task: Implement a linear SVM classifier from scratch using Python and compare its performance with the scikit-learn implementation.

    Use a learning rate and regularization term in the gradient descent process.
    Train the classifier on the training data using the implemented algorithm.
    Implement the predict function to make predictions on new data points.
    Compare the performance of the scratch implementation with scikit-learn's SVM on the Iris dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X, y = iris.data[:, :2], iris.target

# Split the dataset into a training set and a testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# SVM Classifier implementation from scratch
class SVM:
    def __init__(self, learning_rate=0.01, C=1, n_iters=1000):
        self.lr = learning_rate
        self.C = C
        self.n_iters = n_iters
        self.weights = None
        self.bias = None
        self.support_vectors = None

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # Initialize weights and bias
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Gradient Descent optimization
        for _ in range(self.n_iters):
            for i, x in enumerate(X):
                condition = y[i] * (np.dot(x, self.weights) - self.bias) >= 1
                if condition:
                    self.weights -= self.lr * (2 * self.C * self.weights)
                else:
                    self.weights -= self.lr * (2 * self.C * self.weights - np.dot(x, y[i]))
                    self.bias -= self.lr * y[i]
                    if self.support_vectors is None:
                        self.support_vectors = x.reshape(1, -1)
                    else:
                        self.support_vectors = np.vstack((self.support_vectors, x))

    def predict(self, X):
        linear_output = np.dot(X, self.weights) - self.bias
        return np.sign(linear_output)

# Train the SVM classifier from scratch
svm_scratch = SVM(learning_rate=0.01, C=1, n_iters=1000)
svm_scratch.fit(X_train, y_train)

# Predict the labels for the testing set
y_pred_scratch = svm_scratch.predict(X_test)

# Compute the accuracy of the scratch SVM model on the testing set
accuracy_scratch = accuracy_score(y_test, y_pred_scratch)
print(f"Accuracy of SVM classifier from scratch: {accuracy_scratch:.2f}")

# Compare with scikit-learn's SVM
from sklearn.svm import SVC

svm_sklearn = SVC(kernel='linear', C=1)
svm_sklearn.fit(X_train, y_train)
y_pred_sklearn = svm_sklearn.predict(X_test)
accuracy_sklearn = accuracy_score(y_test, y_pred_sklearn)
print(f"Accuracy of scikit-learn's SVM classifier: {accuracy_sklearn:.2f}")

# Plot the decision boundaries of the scratch SVM using two features
def plot_decision_boundary_scratch(svm, X, y):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
    Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Paired)

    # Plot data points
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap=plt.cm.Paired)

    # Mark support vectors
    if svm.support_vectors is not None:
        plt.scatter(svm.support_vectors[:, 0], svm.support_vectors[:, 1],
                    s=100, facecolors='none', edgecolors='k', label='Support Vectors')

    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title('Decision Boundaries of SVM from Scratch')
    plt.legend()
    plt.show()

# Plot decision boundaries of the scratch SVM using two features
# plot_decision_boundary_scratch(svm_scratch, X_train, y_train)
```
