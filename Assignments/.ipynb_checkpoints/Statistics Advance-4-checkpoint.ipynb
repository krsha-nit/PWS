#### Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.
    Ans. 
    T-test:

    The t-test is used when the sample size is small (typically n < 30) and the population standard deviation is unknown.
    It is suitable for testing hypotheses about the means of a single sample or the difference between two independent samples (unpaired samples).
    The t-test assumes that the population follows a normal distribution, but it is robust to violations of this assumption, especially for larger sample sizes.
    Example scenario for a t-test:
    Suppose you want to test if there is a significant difference in the average test scores of two groups of students (Group A and Group B) who received different teaching methods. You would use a t-test to compare the means of the test scores between the two groups and determine if there is a statistically significant difference.

    Z-test:

    The z-test is used when the sample size is large (typically n  30) or when the population standard deviation is known.
    It is appropriate for testing hypotheses about the means of a single sample or the difference between two sample means, assuming a known population standard deviation.
    Example scenario for a z-test:
    Imagine you want to determine if the average weight of a certain population of adults is significantly different from a known population mean weight (e.g., the national average weight). If the sample size is large enough, and you know the population standard deviation (which is rare in practice), you could use a z-test for this comparison.


#### Q2: Differentiate between one-tailed and two-tailed tests.
    Ans. 
    One-tailed test:

    In a one-tailed test, the alternative hypothesis specifies the direction of the effect (e.g., greater than, less than).
    The critical region for the test is on one side of the sampling distribution (either the left tail or the right tail).
    One-tailed tests are used when there is a specific hypothesis about the direction of the effect, and you want to determine if the data supports that direction.
    Example scenario for a one-tailed test:
    You want to test whether a new drug increases the average reaction time compared to a placebo. The one-tailed alternative hypothesis would be that the drug increases reaction time, and you would use a one-tailed test to check if there is evidence to support this claim.

    Two-tailed test:

    In a two-tailed test, the alternative hypothesis does not specify the direction of the effect; it only states that there is a difference or an effect.
    The critical region for the test is split between both tails of the sampling distribution.
    Two-tailed tests are used when you want to determine if there is a significant difference or effect without specifying its direction.
    Example scenario for a two-tailed test:
    Suppose you want to investigate if there is a significant difference in the average scores of a cognitive test between two groups of individuals. The two-tailed alternative hypothesis would simply state that the groups' average scores are different, without indicating whether one group is expected to perform better or worse.


#### Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.
    Ans. 
    Type 1 error (False Positive):

    A Type 1 error occurs when you reject the null hypothesis when it is actually true.
    It represents a false positive result, indicating that you have found a significant effect or relationship when, in reality, it does not exist.
    The probability of committing a Type 1 error is denoted by alpha (), which is the significance level you choose for your hypothesis test.
    Example scenario for a Type 1 error:
    Imagine a medical test to determine if a patient has a particular disease. The null hypothesis is that the patient is healthy. However, due to a false positive result, the test incorrectly indicates that the patient has the disease when, in fact, they do not.

    Type 2 error (False Negative):

    A Type 2 error occurs when you fail to reject the null hypothesis when it is actually false.
    It represents a false negative result, indicating that you have not found a significant effect or relationship when, in reality, it does exist.
    The probability of committing a Type 2 error is denoted by beta ().
    Example scenario for a Type 2 error:
    Continuing with the medical test example, the null hypothesis is again that the patient is healthy. However, due to a false negative result, the test fails to detect the disease when the patient is actually afflicted with it.

#### Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
    Ans. A confidence interval is a range of values that provides an estimate of the true value of a population parameter, such as the mean or proportion, along with a level of confidence in that estimate. It is often used in statistics to quantify the uncertainty associated with sample estimates. The confidence interval indicates the range within which the true population parameter is likely to lie.

    For example, if we calculate a 95% confidence interval for the mean, it means that we are 95% confident that the true population mean falls within that interval.

    Calculating a Confidence Interval (for the mean):
    To calculate a confidence interval for the mean, you'll need the following information:
        The sample mean (X) - the average value of your sample.
        The sample standard deviation (s) - a measure of how much the data points deviate from the sample mean.
        The sample size (n) - the number of observations in your sample.
        The desired confidence level - typically expressed as a percentage (e.g., 90%, 95%, 99%).

    Confidence Interval= X(Critical Values/sqrt(n))
    Where:
    X is the sample mean.
    s is the sample standard deviation.
    n is the sample size.
    Critical Value depends on the desired confidence level and the distribution of the data. For large sample sizes (usually n30), the critical value is taken from the standard normal distribution (Z-distribution). For smaller sample sizes, a t-distribution critical value is used.
    
    Example:

#### Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where a larger sample size would result in a smaller margin of error.
    Ans. 
    Margin of Error:
    The margin of error is a measure of the uncertainty or variability in the estimates obtained from a sample when constructing a confidence interval for a population parameter (e.g., mean, proportion). It indicates the range within which the true population parameter is likely to lie.

    Effect of Sample Size on Margin of Error:
    The margin of error is inversely related to the sample size. In other words, as the sample size increases, the margin of error decreases. A larger sample size provides more information about the population and reduces the variability in the estimates, leading to a more precise confidence interval.

    Example Scenario:
    Suppose you want to estimate the average age of voters in a city. You collect two samples: one with 100 voters and another with 500 voters. Both samples yield the same sample mean of 45 years. Now, you construct 95% confidence intervals for the true average age of voters in the city using both samples.

    For the 100-voter sample, the margin of error might be 2 years.
    For the 500-voter sample, the margin of error might be 1 year.
    In this example, the larger sample size (500 voters) resulted in a smaller margin of error (1 year), providing a more precise estimate of the population's average age.
