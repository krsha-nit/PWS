#### Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose you have collected data on the amount of time students spend studying for an exam and their final exam scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.
    Ans. To calculate the Pearson correlation coefficient between the amount of time students spend studying for an exam (X) and their final exam scores (Y), follow these steps:

    Step 1: Calculate the mean (average) of both variables.
    Step 2: Calculate the standard deviation of both variables.
    Step 3: Calculate the covariance between the two variables.
    Step 4: Divide the covariance by the product of the standard deviations of X and Y.

    Interpretation:
    The Pearson correlation coefficient (r) ranges from -1 to 1.

    If r is close to 1, it indicates a strong positive linear relationship, meaning as the amount of time spent studying increases, the final exam scores tend to increase as well.
    If r is close to -1, it indicates a strong negative linear relationship, meaning as the amount of time spent studying increases, the final exam scores tend to decrease.
    If r is close to 0, it indicates a weak or no linear relationship between the two variables.


#### Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables. Suppose you have collected data on the amount of sleep individuals get each night and their overall job satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two variables and interpret the result.
    Ans. To calculate the Spearman's rank correlation between the amount of sleep individuals get each night (X) and their overall job satisfaction level (Y), follow these steps:

    Step 1: Rank the values of each variable separately. If there are ties (i.e., repeated values), assign the average rank to those tied values.
    Step 2: Calculate the difference between the ranks for each pair of data points (d).
    Step 3: Calculate the Spearman's rank correlation using the formula: rs = 1 - (6 * Σd^2) / (n * (n^2 - 1)), where Σd^2 is the sum of the squared differences between ranks, and n is the number of data points.

    Interpretation:
    Spearman's rank correlation coefficient (rs) ranges from -1 to 1.

    If rs is close to 1, it indicates a strong positive monotonic relationship, meaning as the amount of sleep individuals get each night increases, their overall job satisfaction tends to increase as well.
    If rs is close to -1, it indicates a strong negative monotonic relationship, meaning as the amount of sleep individuals get each night increases, their overall job satisfaction tends to decrease.
    If rs is close to 0, it indicates a weak or no monotonic relationship between the two variables.


#### Q3. Suppose you are conducting a study to examine the relationship between the number of hours of exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation between these two variables and compare the results.
    Ans. To compare the Pearson correlation coefficient and the Spearman's rank correlation between the number of hours of exercise per week (X) and body mass index (BMI) in a sample of 50 adults, follow the same steps as mentioned in Q1 and Q2 for calculating the two correlation coefficients.

    Interpretation:
    The Pearson correlation coefficient measures the strength of the linear relationship between two variables, while the Spearman's rank correlation measures the strength of the monotonic relationship (whether increasing or decreasing) between the two variables.

    If the Pearson correlation coefficient is significantly different from zero, it indicates a significant linear relationship. For example, if it's positive and close to 1, it means that as the number of hours of exercise per week increases, the BMI tends to decrease linearly, and vice versa.
    If the Spearman's rank correlation is significantly different from zero, it indicates a significant monotonic relationship. For example, if it's positive and close to 1, it means that as the number of hours of exercise per week increases, the BMI tends to decrease monotonically, and vice versa.
    Comparing the two correlation coefficients can provide insights into the nature of the relationship between the variables. If both coefficients are similar, it suggests a linear relationship. If the Spearman's rank correlation is higher than the Pearson correlation coefficient, it indicates a monotonic relationship that might not be strictly linear.



#### Q4. A researcher is interested in examining the relationship between the number of hours individuals spend watching television per day and their level of physical activity. The researcher collected data on both variables from a sample of 50 participants. Calculate the Pearson correlation coefficient between these two variables.
    Ans. To calculate the Pearson correlation coefficient between the number of hours individuals spend watching television per day (X) and their level of physical activity (Y) in a sample of 50 participants, follow the steps mentioned in Q1 for calculating the Pearson correlation coefficient.

    Interpretation:
    The Pearson correlation coefficient (r) ranges from -1 to 1.

    If r is close to 1, it indicates a strong positive linear relationship, meaning as the number of hours spent watching television per day increases, the level of physical activity tends to increase as well.

#### Q5. A survey was conducted to examine the relationship between age and preference for a particular brand of soft drink. The survey results are shown below:
    | Age(Years) | Soft drink Preference |
    |------------|----------------------|
    | 25         | Coke                 |
    | 42         | Pepsi                |
    | 37         | Mountain dew         |
    | 19         | Coke                 |
    | 31         | Pepsi                |
    | 28         | Coke                 |

    Ans. To examine the relationship between age and preference for a particular brand of soft drink, you can use a Chi-square test of independence. Since the data is categorical (age groups and soft drink preferences), this test can determine whether there is a significant association between the two variables.

    First, you need to create a contingency table based on the survey results:

    Age (Years)	Coke	Pepsi	Mountain Dew
    25	1	0	0
    42	0	1	0
    37	0	0	1
    19	1	0	0
    31	0	1	0
    28	1	0	0
    Next, perform the Chi-square test of independence on this contingency table to determine if there is a significant association between age and soft drink preference.

    Interpretation:
